```python
import csv
import pandas as pd
from crewai import Agent, Task, Crew
```

```python
def clean_text(file_path):
    try:
        with open(file_path, "r") as file:
            cleaned_lines = [
                line.strip().lower()
                for line in file
                if line.strip()
            ]
        return cleaned_lines
    except FileNotFoundError:
        print("Error: Text file not found.")
        return []
```

```python
text_data = clean_text("data.txt")
text_data[:5]
```

```python
def clean_csv(file_path):
    try:
        df = pd.read_csv(file_path)
        df.dropna(inplace=True)
        return df
    except FileNotFoundError:
        print("Error: CSV file not found.")
        return pd.DataFrame()
```

```python
df = clean_csv("/content/sample_data/california_housing_train.csv")
df.head()
```

```python
import os
from crewai import Agent, Task, Crew
from google.colab import userdata

# Retrieve the API key securely from Colab secrets
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')

agent = Agent(
    role="Data Analyst",
    goal="Analyze cleaned text and CSV data and summarize insights",
    backstory="You are a skilled Data Analyst specializing in extracting key trends and insights from diverse datasets, including both structured CSV data and unstructured text."
)
```

```python
display(df.describe())
```

```python
task = Task(
description="Summarize trends and insights from cleaned datasets",
agent=agent,
expected_output="A comprehensive summary of key trends and insights from both the text data and the CSV data, highlighting any correlations or interesting observations."
)
crew = Crew(
agents=[agent],
tasks=[task]
)
crew.kickoff()
```
